```python
# 컬럼명과 순서가 같은 파일끼리 합치기ㅜ
import pandas as pd
import glob
try:
    path = './dataset/Electrocity/'
    files = glob.glob(path + "*.xlsx")
    excel1 = pd.DataFrame()
    excel2 = pd.DataFrame()
    excel3 = pd.DataFrame()
    excel4 = pd.DataFrame()
    excel5 = pd.DataFrame()
    excel6 = pd.DataFrame()
    for file_name in files:
        df = pd.read_excel(file_name)
        if '행정동코드' in df.columns:
            if '고객호수' in df.columns:
                excel3 = excel3.append(df, ignore_index=True)
            elif '전력 사용량' in df.columns:
                excel5 = excel5.append(df, ignore_index=True)
            else:
                excel1 = excel1.append(df, ignore_index=True)
        elif '읍면동(행정동)' in df.columns:
            if '고객호수' in df.columns:
                excel4 = excel4.append(df, ignore_index=True)
            elif '전력 사용량' in df.columns:
                excel6 = excel6.append(df, ignore_index=True)
            else:
                excel2 = excel2.append(df, ignore_index=True)
            
    #print(excel1)
    #print('='*50)
    #print(excel2)

except Exception as ex:
    print('error' + str(ex))
```
