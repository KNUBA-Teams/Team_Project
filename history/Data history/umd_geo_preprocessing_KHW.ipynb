### 1. 읍면동 geo 데이터 전처리

```python
import pandas as pd

# 데이터 불러오기
df_region = pd.read_csv('../../dataset/geo_hangjeongdong.csv', encoding='utf-8')

df_region
```

```python
# 필요한 column만 저장
df_region = df_region.loc[df_region['sidonm'] == '대구광역시', ['adm_nm','sidonm','temp','sggnm']]
```

```python
# 읍면동을 분리한 column 생성
df_region['emd'] = df_region['temp'].map(lambda x : x.split(' ')[1])

df_region
```

```python
# 전처리 완료 데이터 저장
df_region.to_csv('./output/Daegu_geo_emd.csv', index=False)
```

```python
# 인구밀도 데이터 불러오기
df_density = pd.read_csv('../../output/Daegu_density_final.csv', index_col=0)

df_density
```

```python
# 읍면동 데이터 불러오기
df_region = pd.read_csv('./output/Daegu_geo_emd.csv')

df_region
```

```python
# 읍면동 데이터에 맞춰 인구밀도 데이터를 수정
df_density['region'] = df_density['region'].replace('신천1.2동','신천1·2동')
df_density['region'] = df_density['region'].replace('수성2.3가동','수성2·3가동')
df_density['region'] = df_density['region'].replace('비산2.3동','비산2·3동')
df_density['region'] = df_density['region'].replace('내당2.3동','내당2·3동')
df_density['region'] = df_density['region'].replace('불로.봉무동','불로·봉무동')
df_density['region'] = df_density['region'].replace('두류1.2동','두류1,2동')
```

##### 유천동 처리

2021년 11월 29일 행정동 진천동과 월성1동에서 유천동으로 분동  
출처: https://ko.wikipedia.org/wiki/%EC%9C%A0%EC%B2%9C%EB%8F%99_(%EB%8C%80%EA%B5%AC)#cite_note-9

```python
# 유천동의 분동 기준이 명확하지 않으므로 제거
df_region.drop(df_region.loc[df_region['emd']=='유천동'].index, inplace=True)
```

##### 안심3.4동 처리

2020년 7월부터 안심3.4동이 안심 3동과 4동, 혁신동으로 분리되었다.  
출처 : http://news.imaeil.com/page/view/2020060416223721765

```python
# 안심3.4동 확인
df_density[df_density['region']== '안심3.4동'].tail(5)
```

```python
# 안심3동 확인
df_density[df_density['region']== '안심3동'].head(2)
```

```python
# 안심4동 확인
df_density[df_density['region']== '안심4동'].head(2)
```

```python
# 혁신동 확인
df_density[df_density['region']== '혁신동'].head(2)
```

```python
# 2020년 7월 이전의 결측값 채우기

# 동별 인구수 평균값
ansim3_count = df_density.loc[df_density['region']== '안심3동', 'count'].mean()
ansim4_count = df_density.loc[df_density['region']== '안심4동', 'count'].mean()
hs_count = df_density.loc[df_density['region']== '혁신동', 'count'].mean()

for year in range(2010, 2021):

    if year == 2010:
        for month in range(10, 13):

            # 열지수 (해당 년월의 안심3.4동 열지수)
            ansim_HI = df_density.loc[(df_density['year'] == year) & (df_density['month'] == month) & (df_density['region'] == '안심3.4동'), 'HI'].values[0]

            df_density.loc[len(df_density)] = [year, month, ansim3_count, 6.1, '안심3동', ansim3_count/6.1, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, ansim4_count, 2.53, '안심4동', ansim4_count/2.53, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, hs_count, 16.42, '혁신동', hs_count/16.42, ansim_HI]
    
    elif year == 2020:
        for month in range(1, 7):

            # 열지수 (해당 년월의 안심3.4동 열지수)
            ansim_HI = df_density.loc[(df_density['year'] == year) & (df_density['month'] == month) & (df_density['region'] == '안심3.4동'), 'HI'].values[0]

            df_density.loc[len(df_density)] = [year, month, ansim3_count, 6.1, '안심3동', ansim3_count/6.1, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, ansim4_count, 2.53, '안심4동', ansim4_count/2.53, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, hs_count, 16.42, '혁신동', hs_count/16.42, ansim_HI]
    
    else:
        for month in range(1, 13):

            # 열지수 (해당 년월의 안심3.4동 열지수)
            ansim_HI = df_density.loc[(df_density['year'] == year) & (df_density['month'] == month) & (df_density['region'] == '안심3.4동'), 'HI'].values[0]

            df_density.loc[len(df_density)] = [year, month, ansim3_count, 6.1, '안심3동', ansim3_count/6.1, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, ansim4_count, 2.53, '안심4동', ansim4_count/2.53, ansim_HI]
            df_density.loc[len(df_density)] = [year, month, hs_count, 16.42, '혁신동', hs_count/16.42, ansim_HI]
```

```python
# 안심3.4동 제거
df_density.drop(df_density.loc[df_density['region']=='안심3.4동'].index, inplace=True)
```

```python
# 인구밀도 데이터에만 있는 읍면동 재확인
set(df_density['region'].unique()) - set(df_region['emd'].unique())
```

### 3. 전처리 완료 데이터 조인

```python
df_join = pd.merge(df_density, df_region, left_on=['region'], right_on = ['emd'], how='inner')

df_join
```

```python
# 중복값 제거
df_join.drop(columns='region', inplace=True)
```

```python
# null값 확인
df_join.isnull().sum()
```

```python
# 저장
df_join.to_csv('./output/Daegu_density_for_tableau.csv', index=False)
```
