* tiny data로 처리한 df를 concat하는 작업

```python
# 앞서 정의했던 지역 리스트
regions = ['중구','동구','서구','남구','북구','달서구','달성군','수성구']

# 생성된 df들을 concat하기 위한 리스트
df_concat_list = []

for region in regions:

    # 지역별 파일명을 추출
    files_joined = os.path.join('./dataset/pop_density/population_2', f"*_{region}.csv")

    # 조인된 파일의 리스트를 반환
    list_files = glob.glob(files_joined)
    # print(list_files)

    # 파일 리스트를 전부 조회
    for year in range(2010, 2023):

            df_test = pd.read_csv(f'{list_files[year-2010]}', encoding='utf-8')

            # 2010년의 경우 10월~12월 데이터만 있어 분리
            if year == 2010:

                for month in range(10, 13):
                    
                    # 월별로 처리하기 위해 df를 column별로 분리
                    df_temp = df_test[['행정구역',f'{year}-{month}']]

                    # year와 month 컬럼 추가
                    df_temp['month'] = df_temp.columns[1][5:]
                    df_temp['year'] = df_temp.columns[1][:4]

                    # 인구수 value가 있던 column명을 count로 변경 
                    df_temp.rename(columns={f'{year}-{month}': 'count'}, inplace=True)

                    # concat할 리스트에 추가
                    df_concat_list.append(df_temp)

            # 2022년의 경우 1월~4월 데이터만 있어 분리
            elif year == 2022:

                for month in range(1, 5):

                    # 월별로 처리하기 위해 df를 column별로 분리
                    df_temp = df_test[['행정구역',f'{year}-{month}']]

                    # year와 month 컬럼 추가
                    df_temp['month'] = df_temp.columns[1][5:]
                    df_temp['year'] = df_temp.columns[1][:4]

                    # 인구수 value가 있던 column명을 count로 변경
                    df_temp.rename(columns={f'{year}-{month}': 'count'}, inplace=True)

                    # concat할 리스트에 추가
                    df_concat_list.append(df_temp)
            
            # 2010년과 2022년을 제외한 데이터
            else:
                
                for month in range(1, 13):

                    # 월별로 처리하기 위해 df를 column별로 분리
                    df_temp = df_test[['행정구역',f'{year}-{month}']]

                    # year와 month 컬럼 추가
                    df_temp['month'] = df_temp.columns[1][5:]
                    df_temp['year'] = df_temp.columns[1][:4]

                    # 인구수 value가 있던 column명을 count로 변경
                    df_temp.rename(columns={f'{year}-{month}': 'count'}, inplace=True)

                    # concat할 리스트에 추가
                    df_concat_list.append(df_temp)


concated_df = pd.concat(df_concat_list)

concated_df = concated_df.reset_index(drop=True)

# column 순서 변경
concated_df = concated_df[['행정구역', 'year', 'month', 'count']]

concated_df.to_csv(f'./dataset/pop_density/Deagu_population.csv', mode='w')
```
