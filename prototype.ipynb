```python
x = df[['year','month','day','hour','temp','rainfall','humidity','wind_speed','wind_direction','region','HI','DI']]
y = df[next_temp_list]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
# 모델을 신경망 구조로 설정
model = Sequential()

# 첫번째 Layer, 입력층, 12개를 입력받아 12개를 출력
model.add(Dense(12, input_dim=x_train.shape[1], activation = None))
# 두번째 Layer, 은닉층, 12개를 입력받아 12개를 출력
model.add(Dense(12, activation = None))
# 세번째 Layer, 은닉층, 12개를 입력받아 50개를 출력
model.add(Dense(50, activation = None))
# 네번째 Layer, 출력층, 50개를 입력받아 72개를 출력
model.add(Dense(72, activation = None))
```

```python
model.summary()
```

```python
# binary_crossentropy : 클래스가 두개인 이진 분류에서 사용하는 손실함수
# mse : 예측시스템에서 주로 사용하는 손실함수 (mean square error)
# adam : 신경망의 가중치를 최적화시켜주는 알고리즘
# compile : 만들어진 layer를 ANN으로 묶어서 처리하는 과정
adam=optimizers.Adam(learning_rate=0.15, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='mae', optimizer=adam, metrics=['mae'])
model.fit(x_train, y_train, epochs=30, batch_size=100)
```

```python
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
