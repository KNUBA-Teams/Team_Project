# 러닝을 위한 데이터 전처리

## 지역 데이터를 숫자로 치환

```python
fact = pd.factorize(df['region'])
```

```python
df['region'] = fact[0]
```

```python
df
```

## target 데이터

```python
df['next_temp'] = -50
df['next_humi'] = -50
l = df.shape[0]
```

```python
pred_time = 48
df.loc[0:l-pred_time-1,'next_temp'] = list(df.loc[1:l-pred_time,'temp'])
df.loc[0:l-pred_time-1,'next_humi'] = list(df.loc[1:l-pred_time,'humidity'])
```

```python
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
del_list = list(df[df['datetime'] >= datetime.datetime(2022,5,1,0) - datetime.timedelta(hours=pred_time)].index)
```

```python
df.drop(del_list, inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

## 자 드가자

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras import optimizers
import numpy as np
```

```python
df
```

```python
x = df[['year','month','day','hour','temp','rainfall','humidity','wind_speed','wind_direction','region','HI','DI']]
y = df[['next_temp', 'next_humi']]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
# 모델을 신경망 구조로 설정
model = Sequential()

# 첫번째 Layer, 입력층, 12개를 입력받아 12개를 출력
model.add(Dense(12, input_dim=x_train.shape[1], activation = None))
# 두번째 Layer, 은닉층, 12개를 입력받아 8개를 출력
model.add(Dense(8, activation = None))
# 세번째 Layer, 출력층, 8개를 입력받아 1개를 출력
model.add(Dense(2, activation = None))
```

```python
# binary_crossentropy : 클래스가 두개인 이진 분류에서 사용하는 손실함수
# mse : 예측시스템에서 주로 사용하는 손실함수 (mean square error)
# adam : 신경망의 가중치를 최적화시켜주는 알고리즘
# compile : 만들어진 layer를 ANN으로 묶어서 처리하는 과정
adam=optimizers.Adam(learning_rate=0.15, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss='mae', optimizer=adam, metrics=['mae'])
model.fit(x_train, y_train, epochs=30, batch_size=100)
```

```python
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = pd.DataFrame(y_pred, columns=['pred_temp','pred_humi'])
```

```python
df_ = x_test[:]
df_ = pd.concat([df_,y_test], axis=1)
df_['pre_temp'] = 0
df_['pre_humi'] = 0

df_.loc[:,'pre_temp'] = list(y_pred['pred_temp'])
df_.loc[:,'pre_humi'] = list(y_pred['pred_humi'])
```
